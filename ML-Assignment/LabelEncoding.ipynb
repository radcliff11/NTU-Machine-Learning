```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('C:\\Users\\byeby\\Desktop\\cz4041\\train.csv')
```

```python
data
```

```python
trainX,trainY = data.iloc[:,0:24],data.iloc[:,-1]
print(trainX.shape[0],trainX.shape[1])
```

```python
trainX
```

```python
trainY
```

```python
trainLabelX = pd.DataFrame()
label=LabelEncoder()
for c in trainX.columns:
    if(trainX[c].dtype == 'object'):
        trainLabelX[c] = label.fit_transform(trainX[c])
    else:
        trainLabelX[c] = trainX[c]
```

```python
from sklearn.linear_model import LogisticRegression as lg
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


X_train,X_test,y_train,y_test=train_test_split(trainLabelX,trainY,random_state=42,test_size=0.2)
lr=lg()
lr.fit(X_train,y_train)
y_pre=lr.predict(X_test)
print(accuracy_score(y_test,y_pre))
```

```python
testX = pd.read_csv('C:\\Users\\byeby\\Desktop\\cz4041\\test.csv')
testXlabel = pd.DataFrame()
for c in testX.columns:
    if testX[c].dtype == 'object':
        testXlabel[c] = label.fit_transform(testX[c])
    else:
        testXlabel[c] = testX[c]
print(testX.shape[0],testX.shape[1])
```

```python
test_target = lr.predict_proba(testXlabel)[:,1]
test_target
```

```python
submission = pd.read_csv('C:\\Users\\byeby\\Desktop\\cz4041\\sample_submission.csv')
submission['target'] = test_target
submission.to_csv('label_model.csv')

submission
```
